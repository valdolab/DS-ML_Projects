# Time series processing

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
```

```python
data_set = pd.read_csv("dataset_day1_1fp2s_mean5min.csv")
data_set['ds'] = pd.to_datetime("08:02")
data_set.head()
data_set['ds'] += pd.TimedeltaIndex(data_set.groupby('ds').cumcount()*5, unit='m')
data_set['ds'] = list(map(lambda x: x.time(), data_set['ds']))
data_set
```

```python
data_set.plot(x="ds", y="ts")
```

## Some pre-processing

```python
smooted_ts = data_set['ts'].rolling(2).sum()
```

```python
smooted_ts.plot(x="ds", y="ts")
```
